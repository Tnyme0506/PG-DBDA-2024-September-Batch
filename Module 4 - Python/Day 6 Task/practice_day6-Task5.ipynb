5) Extract any website data using Beautiful Soup.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_manga'

page = requests.get(url)

soup = BeautifulSoup(page.text,'html')
soup
```

```python
table = soup.find_all('table')[1]
```

```python
table
```

```python
title = table.find_all('th')
title = [t.text.strip() for t in title]
title[-3] = 'No. of collected Volumes'
title
```

```python
df = pd.DataFrame(columns = title)
df
```

```python
column_data = table.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    each_row_data = [t.text.strip() for t in row_data]
    print(each_row_data)
    length = len(df)
    df.loc[length] = each_row_data
```

```python
df
```

```python
df.to_csv('best_manga.csv',index = False)
```
